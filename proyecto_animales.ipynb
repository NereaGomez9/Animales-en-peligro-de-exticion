```python
import pandas as pd

df = pd.read_excel('animales_filtrados.xlsx')

# Normalizar columnas
df.columns = df.columns.str.lower().str.replace(' ', '_')

df.head(10)
```

```python
# Lista de grupos que NO son animales
no_animales = ['planta','plantas vasculares','hongos','briofitos','flora','algas','cromistas y bacterias']

# Filtrar solo animales
df_animales = df[~df['grupo'].str.lower().isin(no_animales)]

# Quitar filas sin nombre científico
df_animales = df_animales.dropna(subset=['nombre_cientifico'])

df_animales.head(10)
```

```python
# Eliminar invertebrados
df_animales = df_animales[df_animales['grupo'].str.lower() != 'invertebrados']

df_animales['grupo'].value_counts()
```

```python
df_animales
```

```python
# Número total de filas
total_filas = len(df_animales)

# Número de nombres científicos únicos
total_unicos = df_animales['nombre_cientifico'].nunique()

print(f"Total de filas: {total_filas}")
print(f"Nombres científicos únicos: {total_unicos}")

# Si hay duplicados
if total_filas > total_unicos:
    print(f"Hay {total_filas - total_unicos} filas duplicadas por nombre científico.")
else:
    print("No hay duplicados en la columna 'nombre_cientifico'.")
```

```python
import re

# Elimina "perdida de habitat" solo cuando esté acompañada de otra categoría
df_animales['categoria_amenaza'] = df_animales['categoria_amenaza'].str.replace(
    r'(^pérdida de hábitat/|/pérdida de hábitat$)',  # al principio, al final o en medio
    '/',
    regex=True
)

# Limpieza final: eliminar barras duplicadas o al inicio/final
df_animales['categoria_amenaza'] = (
    df_animales['categoria_amenaza']
    .str.replace(r'//+', '/', regex=True)
    .str.strip('/')
    .str.strip()
)
```

```python
import matplotlib.pyplot as plt

# Normalizar texto: quitar espacios extra y pasar a minúsculas
df_combinado['estado_conservacion_clean'] = df_combinado['estado_conservacion'].str.strip().str.lower()

# Filtrar filas que NO sean 'no evaluado'
df_filtrado = df_combinado[df_combinado['estado_conservacion_clean'] != 'no evaluado']

# Contar especies por estado de conservación
estado_counts = df_filtrado['estado_conservacion'].value_counts().sort_values(ascending=False)

# Crear gráfico de barras verticales
plt.figure(figsize=(10,6))
ax = estado_counts.plot(kind='bar', color='salmon')
plt.xlabel('Estado de conservación')
plt.ylabel('Número de especies')
plt.title('Número de especies por estado de conservación')
plt.xticks(rotation=45)

# Añadir número de especies encima de cada barra
for i, v in enumerate(estado_counts):
    ax.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.show()
```
