```python
import pandas as pd

# Cargar CSV
df = pd.read_excel('animales_filtrados.xlsx')

# Normalizar columnas
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Ver las primeras filas
df.head(10)
```

```python
# Lista de grupos que NO son animales
no_animales = ['planta','plantas vasculares','hongos','briofitos','flora','algas','cromistas y bacterias']

# Filtrar solo animales
df_animales = df[~df['grupo'].str.lower().isin(no_animales)]

# Quitar filas sin nombre científico
df_animales = df_animales.dropna(subset=['nombre_cientifico'])

# Ver las primeras filas filtradas
df_animales.head(10)
```

```python
# Eliminar invertebrados
df_animales = df_animales[df_animales['grupo'].str.lower() != 'invertebrados']

# Ver resultado
df_animales['grupo'].value_counts()
```

```python
df_animales
```

```python
# Número total de filas
total_filas = len(df_animales)

# Número de nombres científicos únicos
total_unicos = df_animales['nombre_cientifico'].nunique()

print(f"Total de filas: {total_filas}")
print(f"Nombres científicos únicos: {total_unicos}")

# Si hay duplicados
if total_filas > total_unicos:
    print(f"Hay {total_filas - total_unicos} filas duplicadas por nombre científico.")
else:
    print("No hay duplicados en la columna 'nombre_cientifico'.")
```

```python
# Mantener solo la primera aparición de cada nombre científico
df_animales = df_animales.drop_duplicates(subset=['nombre_cientifico'], keep='first')

print(f"Filas después de eliminar duplicados: {len(df_animales)}")
```

```python
import pandas as pd

# Cargar ODS
df_boe = pd.read_excel('boe-animales-conservacion.ods', engine='odf', header=None)

# Eliminar filas completamente vacías
df_boe = df_boe.dropna(how='all').reset_index(drop=True)

# Asignar manualmente columnas que queremos
df_boe.columns = ['nombre_cientifico', 'nombre_comun', 'localizacion', 'estado_conservacion']

# Convertir todo a string, limpiar espacios y minúsculas
for col in df_boe.columns:
    df_boe[col] = df_boe[col].astype(str).str.strip().str.lower()

# Verificar
print(df_boe.head())
print(df_boe.columns.tolist())
```

```python
# Eliminar la fila 0
df_boe = df_boe.drop(index=0).reset_index(drop=True)

# Verificar
print(df_boe.head())
```

```python
import pandas as pd

# Crear diccionario de estado de conservación
conservacion_dict = df_boe.set_index('nombre_cientifico')['estado_conservacion'].to_dict()

# Rellenar df_animales
df_animales['estado_conservacion'] = df_animales['estado_conservacion'].fillna(
    df_animales['nombre_cientifico'].str.lower().str.strip().map(conservacion_dict)
)
```
