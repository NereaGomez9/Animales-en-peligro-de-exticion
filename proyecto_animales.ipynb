```python
import pandas as pd

# Cargar CSV
df = pd.read_excel('animales_filtrados.xlsx')

# Normalizar columnas
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Ver las primeras filas
df.head(10)
```

```python
# Lista de grupos que NO son animales
no_animales = ['planta','plantas vasculares','hongos','briofitos','flora','algas','cromistas y bacterias']

# Filtrar solo animales
df_animales = df[~df['grupo'].str.lower().isin(no_animales)]

# Quitar filas sin nombre científico
df_animales = df_animales.dropna(subset=['nombre_cientifico'])

# Ver las primeras filas filtradas
df_animales.head(10)
```

```python
# Eliminar invertebrados
df_animales = df_animales[df_animales['grupo'].str.lower() != 'invertebrados']

# Ver resultado
df_animales['grupo'].value_counts()
```

```python
df_animales
```

```python
# Número total de filas
total_filas = len(df_animales)

# Número de nombres científicos únicos
total_unicos = df_animales['nombre_cientifico'].nunique()

print(f"Total de filas: {total_filas}")
print(f"Nombres científicos únicos: {total_unicos}")

# Si hay duplicados
if total_filas > total_unicos:
    print(f"Hay {total_filas - total_unicos} filas duplicadas por nombre científico.")
else:
    print("No hay duplicados en la columna 'nombre_cientifico'.")
```

```python
# Mantener solo la primera aparición de cada nombre científico
df_animales = df_animales.drop_duplicates(subset=['nombre_cientifico'], keep='first')

print(f"Filas después de eliminar duplicados: {len(df_animales)}")
```

```python
import pandas as pd

# Cargar tu dataset de 4000 especies
df_animales = pd.read_excel("animales_filtrados.xlsx")  # o .xlsx
df_animales['nombre_cientifico'] = df_animales['nombre_cientifico'].str.strip().str.lower()

# Lista de especies en peligro (nombres científicos)
especies_peligro = [
    "lynx pardinus",
    "ursus arctos arctos",
    "mustela lutreola",
    "galemys pyrenaicus",
    "monachus monachus",
    "myotis capaccini",
    "ammotragus lervia",
    "ovis orientalis mussimon",
    "tetrao urogallus cantabricus",
    "aquila adalberti",
    "gypaetus barbatus",
    "ciconia nigra",
    "oxyura leucocephala",
    "neophron percnopterus",
    "chlamydotis undulata fuertaventurae",
    "anser brachyrhynchus",
    "vipera latasti",
    "alytes muletensis",
    "calotriton arnoldi",
    "pelodytes punctatus"
]

# Normalizar nombres
especies_peligro = [x.strip().lower() for x in especies_peligro]

# Crear columna de estado de conservación
df_animales['estado_conservacion'] = df_animales['nombre_cientifico'].apply(
    lambda x: "Peligro de Extinción" if x in especies_peligro else ""
)

# Guardar el resultado
df_animales.to_excel("animales_con_estado.xlsx", index=False)
```
